```python
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
```

```python
fleurs_snr_paths = glob('../data/fleurs/*snr.csv')
fleurs_snr_paths
```

```python
tira_snr_path = '../data/tira-asr-hf/metadata-combined.csv'
snr_paths = fleurs_snr_paths + [tira_snr_path,]
snr_paths
```

```python
fig, axes = plt.subplots(ncols=len(snr_paths),figsize=(12,6))
for i, snr_csv in enumerate(snr_paths):
    df=pd.read_csv(snr_csv)
    if 'tira' in snr_csv:
        df=df.sort_values('nist_stnr', ascending=False)
    x=df.iloc[:1000].boxplot('nist_stnr', ax=axes[i])
    axes[i].set_title(snr_csv)
    axes[i].set_ylim(0,100)
```
