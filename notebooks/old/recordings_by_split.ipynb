```python
import pandas as pd
import os
```

```python
metadata='/Users/markjos/projects/malachor5/data/tira-clean/metadata.csv'
crosstab='/Users/markjos/projects/malachor5/data/tira-clean/rows_per_split.csv'

df=pd.read_csv(metadata)
```

```python
df['eaf_source']=df['eaf_source'].apply(os.path.basename)
eaf_split_crosstab=pd.crosstab(df['eaf_source'], df['split'])
eaf_split_crosstab
```

```python
eaf_split_crosstab=eaf_split_crosstab.sort_values('train')
eaf_split_crosstab
```

```python
eaf_split_crosstab.to_csv(crosstab)
```
