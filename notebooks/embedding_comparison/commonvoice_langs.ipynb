```python
with open(language_codes_path) as f:
    lang_codes = json.load(f)
lang_codes
```

```python
print(len(lang_codes))
for code, name in LANGUAGES.items():
    new_dict={'commonvoice_code': code, 'commonvoice_name': name}
    try:
        iso2 = str(langcodes.find(name))
    except:
        iso2 = code
    for lang in lang_codes:
        if lang.get('iso2', lang.get('commonvoice_code', None))==iso2:
            lang.update(**new_dict)
            break
    else:
        lang_codes.append(new_dict)
len(lang_codes)
```
