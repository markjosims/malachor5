# Beam search
Sandbox for playing around with beam search using HuggingFace `transformers`

```python
from transformers import WhisperForConditionalGeneration, WhisperProcessor
import torchaudio
import torch
```

```python
model=WhisperForConditionalGeneration.from_pretrained('openai/whisper-tiny')
processor=WhisperProcessor.from_pretrained('openai/whisper-tiny')
```

```python
decoder_ids=processor.get_decoder_prompt_ids(task='transcribe', language='english')
```

```python
wav_path='/Users/markjos/projects/malachor5/data/tira-snippets-30s/clips/eng/HH01082021-m00s14ms527-m00s29ms191.wav'
wav, sr=torchaudio.load(wav_path)
wav.shape, sr
```

```python
model_input=processor(wav.squeeze(), sampling_rate=sr, return_tensors='pt').input_features
model_input
```

```python
model.eval()
model_out=model(model_input, decoder_input_ids=torch.tensor(decoder_ids))
model_out
```

```python
model.generation_config.num_beams=4
```
